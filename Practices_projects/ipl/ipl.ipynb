```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x:x.split('--')[1])
df['highest_wickets'] = df['highest_wickets'].astype(int)
bowler= df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
bowler
```

```python
sns.violinplot(y= bowler.index,x = bowler.values,palette = 'rainbow')
```
